```r
plotDifferentialCellTypes <- function(data, cancer_type, output_dir) {
    
    allcomp <- c()
    cell <- unique(data$cell_type)
    
    #run two-sided wilcox test and compute mean foldchange for each cell type
    for(i in 1:length(cell)) {
    df <- subset(data, cell_type %in% cell[i])
    pheno.counts <- as.data.table(table(df$status))
    labels <- paste0(pheno.counts[1,1],".n.", pheno.counts[1,2], ".vrs.", pheno.counts[2,1], ".n.", pheno.counts[2,2])
    df.wilcox <- wilcox_test(df, rank ~ status, ref.group = pheno.counts[2,1]$V1)
    df.mean <- df %>% group_by(status) %>% summarize(Mean = mean(rank, na.rm=TRUE)) %>% as.data.frame()
    df.mean.FC <- df.mean$Mean[2]/df.mean$Mean[1] %>% as.data.frame()
    colnames(df.mean.FC) <- c("foldchange")
    df.mean.FC$cell_type <- cell[i]
    df.all <- cbind(df.wilcox, df.mean.FC)
    allcomp <- rbind(allcomp, df.all)
    }
    
    #multiple testing correction
    allcomp$p.adj <- p.adjust(allcomp$p, method="fdr")
    
    
    #order by p.adj
    allcomp <- allcomp[order(allcomp[,10]),]
    
    
    #create color for volcano plot
    allcomp$diffranked <- "Not Sig"
    allcomp$diffranked[allcomp$foldchange > 1 & allcomp$p.adj < 0.05] <- "Up"
    allcomp$diffranked[allcomp$foldchange < 1 & allcomp$p.adj < 0.05] <- "Down"

    
    allcomp$delabel <- NA
    allcomp[allcomp$diffranked == "Down" | allcomp$diffranked == "Up",]$delabel <- allcomp[allcomp$diffranked == "Down" | allcomp$diffranked == "Up",]$cell_type

    # add compartment and cell_type_tissue
    allcomp$cell_type <- str_trim(allcomp$cell_type)
    allcomp <- allcomp %>% left_join(compartment_map, by = c("cell_type" = "cell_type"))
    allcomp <- allcomp %>%
    # Create a new column 'top_label' based on the conditions
    mutate(
    top_label = case_when(
      diffranked == "Down" & foldchange %in% head(sort(foldchange[diffranked == "Down"], decreasing = FALSE), 10) ~ cell_type,  # Top 10 lowest foldchange for "Down"
      diffranked == "Up" & foldchange %in% head(sort(foldchange[diffranked == "Up"], decreasing = TRUE), 10) ~ cell_type,    # Top 10 highest foldchange for "Up"
      TRUE ~ NA_character_  # Set NA for all other rows
    )
    )

    line_up_fc_value <- min(allcomp %>% filter(diffranked =="Up", !is.na(top_label)) %>% pull(foldchange))
    line_down_fc_value <- max(allcomp%>% filter(diffranked =="Down", !is.na(top_label)) %>% pull(foldchange))
    
    p_sgnf_base_line <- min(allcomp %>% filter(diffranked == "Not Sig") %>% pull(p))

    xlim_up <- max(allcomp %>% pull(foldchange))
    xlim_up <- xlim_up * 1.3

    xlim_down <- min(allcomp %>% pull(foldchange))
    xlim_down <- xlim_down * 0.7
    
    #set colors for volcano plot
    mycolors <- c("#541352FF", "#10a53dFF", "grey80")
    names(mycolors) <- c("Down", "Up", "Not Sig")

    
    my_compartment_shapes <- c(
      "immune" = 15,       
      "endothelial" = 16,  
      "epithelial" = 17,  
      "stromal" = 3,      
      "unknown" = 18       
    )


    # plot <- ggplot(data=allcomp, aes(x=log2(foldchange), y=-log10(p), color=diffranked, shape=compartment, label=top_label)) +
    #     geom_point() +
    #     theme_minimal() +
        
    #     geom_label_repel(aes(label = ifelse(log2(foldchange) > log2(line_down_fc_value) | 
    #                                       log2(foldchange) < log2(line_up_fc_value), top_label, "")), size = 2) +
    #     scale_color_manual(values = mycolors) +
    #     scale_shape_manual(values = my_compartment_shapes) +
    #     geom_vline(xintercept = log2(line_up_fc_value), color = "gray", linetype = "dotted") +   
    #     geom_vline(xintercept = log2(line_down_fc_value), color = "gray", linetype = "dotted") +
    #     geom_hline(yintercept = -log10(p_sgnf_base_line), color = "gray", linetype = "dotted") +
    #     theme_classic() + 
    #     ggtitle(paste(cancer_type, "vs. Controls")) +
    #     theme(plot.title = element_text(hjust = 0.5)) +
    #     guides(color = guide_legend(override.aes = list(shape = 16)))
    plot <- ggplot(data=allcomp, aes(x=log2(foldchange), y=-log10(p), color=diffranked, shape=compartment, label=top_label)) +
        geom_vline(xintercept = log2(line_up_fc_value), color = "gray", linetype = "dotted", size = 0.2) +   
        geom_vline(xintercept = log2(line_down_fc_value), color = "gray", linetype = "dotted", size = 0.2) +
        geom_hline(yintercept = -log10(p_sgnf_base_line), color = "gray", linetype = "dotted", size = 0.2) +
        
        geom_point() +
        geom_label_repel(aes(label = ifelse(log2(foldchange) > log2(line_down_fc_value) | 
                                          log2(foldchange) < log2(line_up_fc_value), top_label, "")), size = 2, show.legend = FALSE) +
        
        scale_color_manual(values = mycolors) +
        scale_shape_manual(values = my_compartment_shapes) +
    
        guides(
            shape = guide_legend(order = 2),
            color = guide_legend(order = 1, override.aes = list(shape = 16, size = 2))  
        ) +
    
        theme_classic() +
        ggtitle(paste(cancer_type, "vs. Controls")) +
        theme(plot.title = element_text(hjust = 0.5, size=16, face = "bold"),
             legend.text = element_text(size = 12),
             legend.title = element_text(size = 13)) +
        xlab("Log2 Fold Change") +
        ylab("-Log10 p-value") +
        scale_x_continuous(limits = c(log2(xlim_down), log2(xlim_up)))


    #ggsave(paste0(output_dir, "/", labels, ".pdf"), height=7, width=7)
    return(list(allcomp = allcomp, plot = plot))
}
```

```r
# lower p value for breast cancer case
plotDifferentialCellTypes_breast <- function(data, output_dir) {
    
    allcomp <- c()
    cell <- unique(data$cell_type)
    
    #run two-sided wilcox test and compute mean foldchange for each cell type
    for(i in 1:length(cell)) {
    df <- subset(data, cell_type %in% cell[i])
    pheno.counts <- as.data.table(table(df$status))
    labels <- paste0(pheno.counts[1,1],".n.", pheno.counts[1,2], ".vrs.", pheno.counts[2,1], ".n.", pheno.counts[2,2])
    df.wilcox <- wilcox_test(df, rank ~ status, ref.group = pheno.counts[2,1]$V1)
    df.mean <- df %>% group_by(status) %>% summarize(Mean = mean(rank, na.rm=TRUE)) %>% as.data.frame()
    df.mean.FC <- df.mean$Mean[2]/df.mean$Mean[1] %>% as.data.frame()
    colnames(df.mean.FC) <- c("foldchange")
    df.mean.FC$cell_type <- cell[i]
    df.all <- cbind(df.wilcox, df.mean.FC)
    allcomp <- rbind(allcomp, df.all)
    }
    
    #multiple testing correction
    #allcomp$p.adj <- p.adjust(allcomp$p, method="fdr")
    allcomp$p.adj <- allcomp$p
    
    
    #order by p.adj
    allcomp <- allcomp[order(allcomp[,10]),]
    
    #export output
    #write.table(allcomp, paste0(output_dir, "/", labels, ".txt"), sep="\t", col.names=T, row.names=F, quote=F)
    
    #create color for volcano plot
    allcomp$diffranked <- "Not Sig"
    allcomp$diffranked[allcomp$foldchange > 1 & allcomp$p.adj < 0.05] <- "Up"
    allcomp$diffranked[allcomp$foldchange < 1 & allcomp$p.adj < 0.05] <- "Down"

    #create color and labels for volcano plot
    allcomp$diffranked_label <- "Not Sig"
    allcomp$diffranked_label[allcomp$foldchange > 1 & allcomp$p.adj < 0.005] <- "Up"
    allcomp$diffranked_label[allcomp$foldchange < 1 & allcomp$p.adj < 0.005] <- "Down"
    
    allcomp$delabel <- NA
    allcomp[allcomp$diffranked == "Down" | allcomp$diffranked == "Up",]$delabel <- allcomp[allcomp$diffranked == "Down" | allcomp$diffranked == "Up",]$cell_type

    allcomp <- allcomp %>%
    # Create a new column 'top_label' based on the conditions
    mutate(
    top_label = case_when(
      diffranked == "Down" & foldchange %in% head(sort(foldchange[diffranked_label == "Down"], decreasing = FALSE), 10) ~ cell_type,  # Top 10 lowest foldchange for "Down"
      diffranked == "Up" & foldchange %in% head(sort(foldchange[diffranked_label == "Up"], decreasing = TRUE), 10) ~ cell_type,    # Top 10 highest foldchange for "Up"
      TRUE ~ NA_character_  # Set NA for all other rows
    )
    )

    line_up_fc <- min(allcomp %>% filter(diffranked_label =="Up") %>%
                    arrange(desc(!is.na(top_label)), top_label) %>%
                    select(diffranked_label, top_label, foldchange) %>%
                    arrange((foldchange)) %>%
                    filter(!is.na(top_label))%>%
                    pull(foldchange))
    line_down_fc <- max(allcomp %>% filter(diffranked_label =="Down") %>%
                    arrange(desc(!is.na(top_label)), top_label) %>%
                    select(diffranked_label, top_label, foldchange) %>%
                    arrange((foldchange)) %>%
                    filter(!is.na(top_label))%>%
                    pull(foldchange))

    #set colors for volcano plot
    mycolors <- c("#541352FF", "#10a53dFF", "grey80")
    names(mycolors) <- c("Down", "Up", "Not Sig")
    
    ### for testing - color according to compartment
    # Load raw correlation data
    compartment_map <- read.csv("/mnt/DATA3/timo/data/compartment_map.csv", row.names = 1)
    removal_words <- c("Small", "Large", "Bone", "Salivary", "Lymph")
    # Modify the tissue_df to remove the last word and second-to-last word if in removal_words
    allcomp <- allcomp %>%
      mutate(cells = case_when(
        str_detect(cell_type, paste0("\\.(", paste(removal_words, collapse = "|"), ")\\.")) ~
          str_remove(cell_type, "\\.[A-Za-z]+\\.[A-Za-z]+$"),  # Remove the last and second-last words
        TRUE ~ str_remove(cell_type, "\\.[A-Za-z]+$")  # Otherwise remove the last word
      ))
    allcomp <- allcomp %>%
    left_join(compartment_map, by = c("cells" = "cells"))
    my_compartment_shapes <- c(
      "immune" = 15,       # Circle
      "endothelial" = 16,  # Triangle
      "epithelial" = 17,   # Diamond
      "stromal" = 3,      # Square
      "unknown" = 18        # Plus
    )


    # volcano plot
    plot <- ggplot(data=allcomp, aes(x=log2(foldchange), y=-log10(p), color=diffranked, shape=compartment, label=top_label)) +
        geom_point() +
        theme_minimal() +
        geom_label_repel(size=2) +
        scale_color_manual(values = mycolors) +
        scale_shape_manual(values = my_compartment_shapes) +
        geom_vline(xintercept = log2(line_up_fc), color = "gray", linetype = "dotted") +   
        geom_vline(xintercept = log2(line_down_fc), color = "gray", linetype = "dotted") + 
        theme_classic() + 
        ggtitle(labels)
    #ggsave(paste0(output_dir, "/", labels, ".pdf"), height=7, width=7)
    return(list(allcomp = allcomp, plot = plot))
}
```
